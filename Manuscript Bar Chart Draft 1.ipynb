```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.ticker as mticker
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
df=pd.read_csv("Manuscript_Barchart.csv")
df.set_index('Registry Names',inplace=True)
fig, (ax1,ax2) = plt.subplots(2,1,sharex=True,
                         figsize=(12,8))


#Spines are the lines connecting the axis tick marks and noting the boundaries of the data area. To turn off the spines of the 
#axis, we run the below 1 code

ax1.spines['bottom'].set_visible(False) 


#Now we want to turn off both major and minor ticks of the axis 'x' of ax1
ax1.tick_params(axis='x',which='both',bottom=False)


#To turn off the spines at the top of the ax2
ax2.spines['top'].set_visible(False)

#We are done with making the canvas. Now, we need to plot our data in the canvas.
#Setting the y-axis numbers and ticks

ax2.set_ylim(bottom=0,top=80)
ax1.set_ylim(bottom=105,top=380)
ax2.set_yticks(np.arange(0,80,5))
ax1.set_yticks(np.arange(105,380,20))


#Setting the color of the bars
my_colors = ['black', 'lightgray']


#Plotting the bars (using reindex such that the registry which has the highest number of records comes first)
#Parameter 'align(Optional)'': Has two possible values, center or edge.
#These values define whether the bars will center their base on the x-coordinate, or 
#align their left edge with the x-coordinate.

df.reindex(["ISRCTN", "ANZCTR", "jRCT","DRKS","ChiCTR","ReBEC","PACTR","IRCT","CRIS","ITMCTR","RPCEC"]).plot(ax=ax1,kind='bar',color=my_colors,width=0.8,align="center")
df.reindex(["ISRCTN", "ANZCTR", "jRCT","DRKS","ChiCTR","ReBEC","PACTR","IRCT","CRIS","ITMCTR","RPCEC"]).plot(ax=ax2,kind='bar',color=my_colors,width=0.8,align="center")



#We need only one legend in the bottom right of ax1. So we turned off the legend of ax2
ax2.legend_ = None


#This code sets the size of the Legend
plt.rc('legend', fontsize=15) 

#We want to annotate the bars
for c in ax1.containers:
    ax1.bar_label(c, label_type='edge')
    
for c in ax2.containers:
    ax2.bar_label(c, label_type='edge')

#Sets the font size for all elements (ticks, labels, title,annotations) in a plot
plt.rcParams.update({'font.size': 11})


#Setting the x-tick-labels rotation to 0 (As default is 90 degree)
for tick in ax2.get_xticklabels():
    tick.set_rotation(0)

# how big to make the diagonal lines in axes coordinates
d = .01


#The below 3 lines creates a diagonal line segment in the subplot ax1 starting from (-0.01, -0.01) 
#and ending at (0.01, 0.01). The line is colored black and extends beyond the axes limits.

#This line creates a dictionary kwargs that contains keyword arguments used for customizing the plot. 
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)

ax1.plot((-d, +d), (-d, +d), **kwargs)      
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)



#The first line segment extends from (-0.01, 0.99) to (0.01, 1.01), 
#and the second line segment extends from (0.99, 0.99) to (1.01, 1.01) relative to the ax2 subplot
#The update() method is called on the kwargs dictionary to add a new key-value pair.

kwargs.update(transform=ax2.transAxes)  
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)


plt.xlabel("Registry names",fontsize=11)
fig.text(0.07, 0.5, 'Number of records', va='center', rotation='vertical',fontsize=11)

fig.suptitle('Comparision of Methodology A and Methodology B', fontsize=13, y=0.95)

plt.show()
fig.savefig("Manuscript_fig_draft1.png")



```

```python
#This script was utilised
df=pd.read_csv("Manuscript_Barchart.csv")
df.set_index('Registry Names',inplace=True)
fig,ax = plt.subplots(figsize=(12,8))
ax.set_ylim(bottom=0,top=370)
ax.set_yticks(np.arange(0,370,20))
my_colors = ['black', 'lightgray']
df.reindex(["ISRCTN", "ANZCTR", "jRCT","DRKS","ChiCTR","ReBEC","PACTR","IRCT","CRIS","ITMCTR","RPCEC"]).plot(ax=ax,kind='bar',color=my_colors,width=0.8,align="center",edgecolor="black")

plt.rc('legend', fontsize=13) 
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1, edgecolor="black")


for c in ax.containers:
    ax.bar_label(c, label_type='edge')
    
for tick in ax.get_xticklabels():
    tick.set_rotation(0)
    
ax.set_ylabel('Number of records', fontsize=11,labelpad=10)
ax.set_xlabel('Registry names', fontsize=11,labelpad=10)
ax.set_title('Comparision of Methodology A and Methodology B', fontsize=13,fontweight ='bold',y=1.02,loc="center")

plt.show()
fig.savefig("Manuscript_fig_draft4.png")


```
